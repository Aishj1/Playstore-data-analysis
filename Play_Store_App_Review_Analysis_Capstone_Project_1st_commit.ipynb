```python
#cleaning the data
#removing insuffiecient data
insuffiecient_data=p_df[p_df['Type']!='Free'][p_df[p_df['Type']!='Free']['Price']=='0']
p_df.drop(insuffiecient_data.index , inplace=True)
```

```python
#removing mismatched values
mismatched_values=p_df[p_df['Genres']== 'February 11, 2018']
p_df.drop(mismatched_values.index, inplace=True)
```

```python
#cleaning strings
def cleaning_data(i):
  '''This functions takes string values and replace them, '+', ',' ,'$', 'M', 'k', 'NaN' '''
  if '+' in i:
    i=i.replace('+', '')
  if ',' in i:
    i=i.replace(',', '')
  if '$' in i:
    i=i.replace('$', '')
  if 'M' in i:
    i=i.replace('M', '')*1000000    
  if 'K' in i:
    i=i.replace('K', '')*1000
  if 'NaN' in i:
    i=0
  else:
    pass
  return i     
```
